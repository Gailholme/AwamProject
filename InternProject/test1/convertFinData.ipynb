# This is to make table beautiful
- assign sheet1 as list to col_names

For loop --> index, row --> index and row of that whole list then you

```python
ALL_SHEETS = x1.sheet_names
for sheet in ALL_SHEETS: # iterator to run all sheet
    if sheet in ['Name', '(54)', '(55)', '(56)', '(57)', '(58)', '(59)', '(60)']:
        continue
    df1= x1.parse(sheet)


    with open('sheet-{}.csv'.format(sheet) , 'w') as csvfile: # open file sheet.csv and write only --> name csv file
        #format sheet --> put in sheet name in {}
        gailwriter = csv.writer(csvfile) #  tell that go write at csv file


        col_names = list(df1)
        DATE_NULL = list(df1.isnull()['Start Date'])
        TICKER_NAME = None

        for index,row in df1.iterrows():
        #     print(index)
            if index in [0,1,3]: # cut the unneceessary
                continue
            date = row[col_names[0]] # date is first column
            if DATE_NULL[index]: # if date is no--> dont run
                continue
            if type(date) is str: # if date is not date ( the name of stock)
                TICKER_NAME = date # save name in Ticker_name
                date = "Date" #set that string to Date
            TICKER_INDEX = DICT_INDEX[TICKER_NAME] # match the ticker name to dictionary

            if type(date) is str:
                gailwriter.writerow((TICKER_INDEX, date, row[col_names[1]], row[col_names[2]], row[col_names[3]], row[col_names[4]] , row[col_names[5]], row[col_names[6]]))
            else:
                gailwriter.writerow((TICKER_INDEX, date.strftime("%m/%Y/%d"), row[col_names[1]], row[col_names[2]], row[col_names[3]], row[col_names[4]] , row[col_names[5]], row[col_names[6]]))
```
